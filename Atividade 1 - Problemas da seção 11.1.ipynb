##### **Questão 1:** Find $\lim_{h \rightarrow 0} \left[ \dfrac{3h + h^2}{h} \right]$.

Por L'Hospital,

$$\lim_{x\rightarrow 0} \dfrac{f(x)}{g(x)} = \lim_{x\rightarrow 0} \dfrac{f'(x)}{g'(x)}$$

Assim,

$$\lim_{h \rightarrow 0} \left[ \dfrac{3h + h^2}{h} \right] = \lim_{h \rightarrow 0} \left[ \dfrac{3 + 2 \cdot h}{1} \right]$$

Portanto,

$$\lim_{h \rightarrow 0} \left[ \dfrac{3h + h^2}{h} \right] = 3$$

##### **Questão 2:** It costs Sugarco 25¢/lb to purchase the first 100 lb of sugar, 20¢/lb to purchase the next 100 lb, and 15¢ to buy each additional pound. Let f(x) be the cost of purchasing x pounds of sugar. Is f(x) continuous at all points? Are there any points where f(x) has no derivative?

A função do problema é denotada por:

$$f(x) = \begin{cases}
    25 \cdot x, & \text{para}\ x \leq 100 \\
    20 \cdot x, & \text{para}\ 100 < x \leq 200 \\
    15 \cdot x, & \text{para}\ x > 200

A função é descontínua em x = 100 e x = 200 e contínua nos outros pontos. Também, não há derivada nesses pontos.

##### **Questão 3:** Find f'(x) for each of the following functions:

a) $x \cdot e^{-x}$

$$f(x) = x \cdot e^{-x}$$

Pela regra do produto, tem-se,

$$f'(x) = g'(x) \cdot h(x) + g(x) \cdot h'(x)$$

No nosso caso,
- $g(x) = x$
- $g'(x) = 1$
- $h(x) = e^{-x}$
- $h'(x) = -e^{-x}$

Portanto,

$$f'(x) = e^{-x} - x \cdot e^{-x}$$

Simplificando,

$$f'(x) = e^{-x} \cdot (1 - x)$$

b) $\dfrac{x^2}{x^2+1}$

$$f(x) = \dfrac{x^2}{x^2+1}$$

Pela regra do quociente,

$$f'(x) = \dfrac{g'(x) \cdot h(x) - g(x) \cdot h'(x)}{h^2(x)}$$

No nosso caso,
- $g(x) = x^2$
- $g'(x) = 2x$
- $h(x) = x^2 + 1$
- $h'(x) = 2x$

Portanto,

$$f'(x) = \dfrac{2x \cdot (x^2+1) - x^2 \cdot 2x}{(x^2+1)^2}$$

Simplificando,

$$f'(x) = \dfrac{2x}{(x^2+1)^2}$$

c) $e^{3x}$

$$f(x) = e^{3x}$$

Pela regra da cadeia,

$$f'(x) = f'(g(x)) \cdot g'(x)$$

No nosso caso,
- $f(g(x)) = e^{3x}$
- $f'(g(x)) = e^{3x}$
- $g(x) = 3x$
- $g'(x) = 3$

Portanto,

$$f'(x) = 3 \cdot e^{3x}$$

d) $(3x+2)^{-2}$

$$f(x) = (3x+2)^{-2}$$

Utilizando as regras anteriores e pela regra de derivação de polinômios,

$$f'(x) = nx^{(n-1)}$$

Portanto,

$$f'(x) = -2 \cdot (3x+2)^{-3} \cdot 3$$

Simplificando,

$$f'(x) = -6 \cdot (3x+2)^{-3}$$

e) $\ln\left(x^3\right)$

$$\ln\left(x^3\right)$$

Utilizando as regras anteriores e pela regra de derivação de logaritmos,

$$f'(x) = \dfrac{1}{x}$$

Portanto,

$$f'(x) = 3\not{x^2} \cdot \dfrac{1}{x^{\not 3}}$$

Simplificando,

$$f'(x) = \dfrac{3}{x}$$

##### **Questão 6:** Let *q=f(p)* be the demand for a product when the price is *p*. For a given price *p*, the price elasticity *E* of the product is defined by $$E = \dfrac{\text{percentage change in demand}}{\text{percentage change in price}}$$ If the change in price ($\Delta p$) is small, this formula reduces to $$E = \dfrac{\frac{\Delta q}{q}}{\frac{\Delta p}{p}} = \left( \dfrac{p}{q} \right) \left( \dfrac{dq}{dp} \right)$$

a) Would you expect *f(p)* to be positive or negative?

É esperado que *f(p)* seja positivo, pois a demanda sempre será positiva.

b) Show that if *E < -1*, a small decrease in price will increase the firm’s total revenue (in this case, we say that demand is *elastic*).

*E < -1* implica que

$$\left( \dfrac{p}{q} \right) \left( \dfrac{dq}{dp} \right) < -1$$

Ou seja,

$$\dfrac{q}{p} < \dfrac{dq}{dp}$$

Assim, a proporção entre a demanda e o preço é menor do que a taxa de variação da demanda em relação ao preço. Por conseguinte, a demanda é bastante sensível às mudanças de preço.

c) Show that if *-1 < E < 0*, a small price decrease will decrease total revenue (in this case, we say demand is *inelastic*).

*-1 < E < 0* implica que

$$-1 < \left( \dfrac{p}{q} \right) \left( \dfrac{dq}{dp} \right) < 0$$

Ou seja,

$$\dfrac{q}{p} > \dfrac{dq}{dp}$$

Assim, a proporção entre a demanda e o preço é maior do que a taxa de variação da demanda em relação ao preço. Por conseguinte, a demanda é pouco sensível às mudanças de preço.

##### **Questão 7:** Suppose that if *x* dollars are spent on advertising during a given year, $k(1 - e^{-cx})$ customers will purchase a product *(c > 0)*.

a) As *x* grows large, the number of customers purchasing the product approaches a limit. Find this limit.

Vamos calcular o limite:

$$\lim_{x \to \infty} k(1-e^{-cx})$$

Como $c>0$, $e^{-cx} = 0$ e então:

$$\lim_{x \to \infty} k(1-e^{-cx}) = k$$

b) Can you give an interpretation for *k*?

*k* representa o número máximo de clientes que podem ser alcançados com a publicidade.

c) Show that the sales response from a dollar of advertising is proportional to the number of potential customers who are not purchasing the product at present.

A quantidade de clientes potenciais ($C_p$) que não estão comprando o produto no momento é dada por:

$$C_p = k - k(1-e^{-cx})$$

Simplificando,

$$C_p = k e^{-cx}$$

Calculando a variação de pontenciais clientes,

$$\dfrac{d C_p}{dx} = -ck \cdot e^{-cx}$$

Nota-se que é proporcional à resposta (variação) do número de clientes ($C$):

$$\dfrac{dC}{dx} = - ck \cdot e^{-cx}$$

##### **Questão 8:** Let the total cost of producing x units, *c(x)*, be given by $c(x) = kx^{1-b} (0 < b < 1)$. This cost curve is called the *learning* or *experience cost curve*.

a) Show that the cost of producing a unit is a decreasing function of the number of units that have been produced.

Derivando $c(x)$, tem-se:

$$\dfrac{dc}{dx} = k \cdot (1-b) \cdot x^{-b}$$

A restrição de *b* é dada por $(0 < b < 1)$, então $k \cdot (1-b) \geq 0\ \forall\ k \geq 0$ e $\frac{dc}{dx} \geq 0$.

Portanto, $c(x)$ é uma função decrescente em *x*.

b) Suppose that each time the number of units produced is doubled, the per-unit product cost drops to *r%* of its previous value (because workers learn how to perform their jobs better). Show that $r = 100(2^{-b})$.

O valor de *r* é dado por:

$$r = \dfrac{\frac{c(2x)}{2x}}{\frac{c(x)}{x}} \cdot 100\%$$

Ou seja,

$$r = \dfrac{\frac{\not{k} \cdot (2x)^{1-b}}{2\not{x}}}{\frac{\not{k} \cdot x^{1-b}}{\not{x}}} \cdot 100\%$$

Simplificando,

$$r = 100 (2^{-b})$$
