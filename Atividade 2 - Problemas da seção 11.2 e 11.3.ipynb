##### **Questão 10:**

For what values of *a*, *b*, and *c* will $ax_1^2 + bx_1x_2 + cx_2^2$ be a convex function on $R^2$? A concave function on $R^2$?

A função é dada por:

$$f(x_1,\ x_2) = ax_1^2 + bx_1x_2 + cx_2^2$$

Para determinar a matriz Hessiana:
- $\dfrac{\partial f}{\partial x_1} = 2ax_1 + bx_2$
- $\dfrac{\partial f}{\partial x_2} = bx_1 + 2cx_2$
- $\dfrac{\partial^2 f}{\partial x_1^2} = 2a$
- $\dfrac{\partial^2 f}{\partial x_1 \partial x_2} = b$
- $\dfrac{\partial^2 f}{\partial x_2^2} = 2c$

$$H(x_1,\ x_2) = \left[\begin{array}{cc}
    \dfrac{\partial^2 f}{\partial x_1^2} & \dfrac{\partial^2 f}{\partial x_1 \partial x_2} \\
    \dfrac{\partial^2 f}{\partial x_1 \partial x_2} & \dfrac{\partial^2 f}{\partial x_2^2}
\end{array}\right] = \left[\begin{array}{cc}
    2a & b \\
    b & 2c
\end{array}\right]$$

$$\det{H} = 4ac - b^2$$

Para que seja **convexa**, $4ac - b^2 > 0$, tem-se que:
- $a \cdot c > 0$
- $-2\sqrt{ac} < b < 2\sqrt{ac}$

Para que seja **côncava**, $4ac - b^2 < 0$, tem-se que:
- $a \cdot c > 0$
- $b > 2\sqrt{ac}$ ou $b < - 2\sqrt{ac}$

##### **Questão 12:**

Show that if $f(x_1,\ x_2,\ \dots,\ x_n)$ and $g(x_1,\ x_2,\ \dots,\ x_n)$ are convex functions on a convex set S, then $h(x_1, x_2,\ \dots, x_n) = f(x_1, x_2,\ \dots, x_n) = g(x_1, x_2,\ \dots, x_n)$ is a convex function on S.

Se *f* e *g* são funções convexas em *S* e *h = f + g*, então utilizando a definição de função convexa dada pela equação $$f[cx'+(1-c)x''] \leq cf(x')+(1-c)f(x''),$$ tem-se que

$$\begin{align}
    h(cx'+(1-c)x'') &= f(cx+(1-c)x'') + g(cx'+(1-c)x'') \\
    &\leq cf(x')+(1-c)f(x'')+cg(x')+(1-c)g(x'') \\
    &= ch(x')+(1-c)h(x'')
\end{align}$$

Então, por definição *h* é convexa.

**Questão 13:**

If $f(x_1,\ x_2,\ \dots,\ x_n)$ is a convex function on a convex set S, show that for $c \geq 0$, $g(x_1,\ x_2,\ \dots,\ x_n) = cf(x_1,\ x_2,\ \dots,\ x_n)$ is a convex function on S, and for $c \leq 0$, $g(x_1,\ x_2,\ \dots,\ x_n) = cf(x_1,\ x_2,\ \dots,\ x_n)$ is a concave function on S.

Se *f* é uma função convexa em *S* e *g = cf*, então utilizando a definição de função convexa dada pela equação $$f[cx'+(1-c)x''] \leq cf(x')+(1-c)f(x''),$$ tem-se que, $\forall x',\ x'' \in S$, $c \geq 0$ e $d \in [0,\ 1]$,

$$\begin{align}
    g(dx'+(1-d)x'') &= cf(dx'+(1-d)x'') \\
    &\leq dcf(x')+(1-d)cf(x'') \\
    &= dg(x')+(1-d)g(x'')
\end{align}$$

Então, por definição *h* é convexa. Por outro lado, se $c \leq 0$ e $d \in [0,\ 1]$,

$$\begin{align}
    g(dx'+(1-d)x'') &= cf(dx'+(1-d)x'') \\
    &\geq dcf(x')+(1-d)cf(x'') \\
    &= dg(x')+(1-d)g(x'')
\end{align}$$

Então, por definição *g* é côncava.
